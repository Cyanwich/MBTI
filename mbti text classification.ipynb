```python
from __future__ import unicode_literals, print_function
import random
from pathlib import Path
import thinc.extra.datasets

import spacy
from spacy.util import minibatch, compounding

from sklearn.model_selection import train_test_split
import pandas as pd
df = pd.read_csv("mbti_prof3.csv")
lm = spacy.load('en')
for x_ in range(len(df[["posts"]])):
    lmtarget = lm(df.iloc[x_,1])
    lmstring = ''
    for word in lmtarget:
        lmstring = lmstring + word.lemma_ + ' '
    df.iloc[x_,1] = lmstring.lower()

def main(model=None, output_dir=None, n_iter=20, n_texts=3000):
    if model is not None:
        nlp = spacy.load(model)  # load existing spaCy model
        print("Loaded model '%s'" % model)
    else:
        nlp = spacy.blank('en')  # create blank Language class
        print("Created blank 'en' model")

    # add the text classifier to the pipeline if it doesn't exist
    # nlp.create_pipe works for built-ins that are registered with spaCy
    if 'textcat' not in nlp.pipe_names:
        textcat = nlp.create_pipe('textcat')
        nlp.add_pipe(textcat, last=True)
    # otherwise, get it, so we can add labels to it
    else:
        textcat = nlp.get_pipe('textcat')

    # add label to text classifier
    textcat.add_label('INTROVERTED')
    textcat.add_label('INTUTIVE')
    textcat.add_label('JUDGEMENTAL')
    textcat.add_label('THINKING')
    textcat.add_label('IN')
    textcat.add_label('IS')
    textcat.add_label('EN')
    textcat.add_label('ES')
    textcat.add_label('IT')
    textcat.add_label('IF')
    textcat.add_label('ET')
    textcat.add_label('EF')
    textcat.add_label('IJ')
    textcat.add_label('IP')
    textcat.add_label('EJ')
    textcat.add_label('EP')
    textcat.add_label('NT')
    textcat.add_label('NF')
    textcat.add_label('ST')
    textcat.add_label('SF')
    textcat.add_label('NJ')
    textcat.add_label('NP')
    textcat.add_label('SJ')
    textcat.add_label('SP')
    textcat.add_label('TJ')
    textcat.add_label('TP')
    textcat.add_label('FJ')
    textcat.add_label('FP')

    # load the IMBD dataset
    print("Loading MBTI data...")
    (train_texts, train_cats), (dev_texts, dev_cats) = load_data(df,limit=n_texts)
    print("Using {} examples ({} training, {} evaluation)"
          .format(n_texts, len(train_texts), len(dev_texts)))
    train_data = list(zip(train_texts,
                          [{'cats': cats} for cats in train_cats]))

    # get names of other pipes to disable them during training
    other_pipes = [pipe for pipe in nlp.pipe_names if pipe != 'textcat']
    with nlp.disable_pipes(*other_pipes):  # only train textcat
        optimizer = nlp.begin_training()
        print("Training the model...")
        print('{:^5}\t{:^5}\t{:^5}\t{:^5}\t{:^5}'.format('ITER','LOSS', 'P', 'R', 'F'))
        for i in range(n_iter):
            losses = {}
            # batch up the examples using spaCy's minibatch
            batches = minibatch(train_data, size=compounding(4., 32., 1.001))
            for batch in batches:
                texts, annotations = zip(*batch)
                nlp.update(texts, annotations, sgd=optimizer, drop=0.2,
                           losses=losses)
            with textcat.model.use_params(optimizer.averages):
                # evaluate on the dev data split off in load_data()
                scores = evaluate(nlp.tokenizer, textcat, dev_texts, dev_cats)
            print('{0:.3f}\t{1:.3f}\t{2:.3f}\t{3:.3f}\t{4:.4f}'  # print a simple table
                  .format(i+1, losses['textcat'], scores['textcat_p'],
                          scores['textcat_r'], scores['textcat_f']))

    # test the trained model
    test_text = "I am an outgoing partier who is a concrete thinker, is in tune with their emotions and feelings, and is pretty indecisive."
    doc = nlp(test_text)
    print(test_text, doc.cats)

    if output_dir is not None:
        output_dir = Path(output_dir)
        if not output_dir.exists():
            output_dir.mkdir()
        nlp.to_disk(output_dir)
        print("Saved model to", output_dir)

        # test the saved model
        print("Loading from", output_dir)
        nlp2 = spacy.load(output_dir)
        doc2 = nlp2(test_text)
        print(test_text, doc2.cats)

def load_data(train_data,limit=0, split=0.4):
    """Load data from the IMDB dataset."""
    # Partition off part of the train data for evaluation
    #random.shuffle(train_data)
    train_data = train_data[-limit:]
    Y,X = train_data["type"], train_data["posts"]
    y = []
    for y_ in Y:
        if y_[0] == 'I' : INTROVERTED = True
        else: INTROVERTED = False
        if y_[1] == 'N' : INTUTIVE=  True
        else: INTUTIVE=  False
        if y_[2] == 'T' : THINKING=  True
        else: THINKING= False
        if y_[3] == 'J' : JUDGEMENTAL=  True
        else: JUDGEMENTAL=  False
            
        if y_[0] == 'I' and y_[1] == 'N' : IN=  True
        else: IN=  False
        if y_[0] == 'I' and y_[1] == 'S' : IS=  True
        else: IS=  False
        if y_[0] == 'E' and y_[1] == 'N' : EN=  True
        else: EN=  False
        if y_[0] == 'E' and y_[1] == 'S' : ES=  True
        else: ES=  False
        if y_[0] == 'I' and y_[2] == 'T' : IT=  True
        else: IT=  False
        if y_[0] == 'I' and y_[2] == 'F' : IF=  True
        else: IF=  False
        if y_[0] == 'E' and y_[2] == 'T' : ET=  True
        else: ET=  False
        if y_[0] == 'E' and y_[2] == 'F' : EF=  True
        else: EF=  False
        if y_[0] == 'I' and y_[3] == 'J' : IJ=  True
        else: IJ=  False
        if y_[0] == 'I' and y_[3] == 'P' : IP=  True
        else: IP=  False
        if y_[0] == 'E' and y_[3] == 'J' : EJ=  True
        else: EJ=  False
        if y_[0] == 'E' and y_[3] == 'P' : EP=  True
        else: EP=  False
        if y_[1] == 'N' and y_[2] == 'T' : NT=  True
        else: NT=  False
        if y_[1] == 'N' and y_[2] == 'F' : NF=  True
        else: NF=  False
        if y_[1] == 'S' and y_[2] == 'T' : ST=  True
        else: ST=  False
        if y_[1] == 'S' and y_[2] == 'F' : SF=  True
        else: SF=  False
        if y_[1] == 'N' and y_[3] == 'J' : NJ=  True
        else: NJ=  False
        if y_[1] == 'N' and y_[3] == 'P' : NP=  True
        else: NP=  False
        if y_[1] == 'S' and y_[3] == 'J' : SJ=  True
        else: SJ=  False
        if y_[1] == 'S' and y_[3] == 'P' : SP=  True
        else: SP=  False
        if y_[2] == 'T' and y_[3] == 'J' : TJ=  True
        else: TJ=  False
        if y_[2] == 'T' and y_[3] == 'P' : TP=  True
        else: TP=  False
        if y_[2] == 'F' and y_[3] == 'J' : FJ=  True
        else: FJ=  False
        if y_[2] == 'F' and y_[3] == 'P' : FP=  True
        else: FP=  False
            
        y.append({'INTROVERTED':INTROVERTED,"INTUTIVE":INTUTIVE,"THINKING":THINKING,"JUDGEMENTAL":JUDGEMENTAL,"IN":IN,"IS":IS,"EN":EN,"ES":ES,"IT":IT,"IF":IF,"ET":ET,"EF":EF,"IJ":IJ,"IP":IP,"EJ":EJ,"EP":EP,"NT":NT,"NF":NF,"ST":ST,"SF":SF,"NJ":NJ,"NP":NP,"SJ":SJ,"SP":SP,"TJ":TJ,"TP":TP,"FJ":FJ,"FP":FP})
            
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=split, random_state=42)
    return (X_train, y_train), (X_test, y_test)

def evaluate(tokenizer, textcat, texts, cats):
    docs = (tokenizer(text) for text in texts)
    tp = 1e-8  # True positives
    fp = 1e-8  # False positives
    fn = 1e-8  # False negatives
    tn = 1e-8  # True negatives
    for i, doc in enumerate(textcat.pipe(docs)):
        gold = cats[i]
        for label, score in doc.cats.items():
            if label in ['INTROVERTED','INTUTIVE','JUDGEMENTAL','THINKING']:
                if score >= 0.5 and gold[label] >= 0.5:
                    tp += 1.
                elif score >= 0.5 and gold[label] < 0.5:
                    fp += 1.
                elif score < 0.5 and gold[label] < 0.5:
                    tn += 1
                elif score < 0.5 and gold[label] >= 0.5:
                    fn += 1
            elif label in ['EN','IS','ES','IN']:
                if score == max([doc.cats['EN'],doc.cats['IS'],doc.cats['ES'],doc.cats['IN']]) and gold[label] >= 0.5:
                    tp += 1.
                elif score == max([doc.cats['EN'],doc.cats['IS'],doc.cats['ES'],doc.cats['IN']]) and gold[label] < 0.5:
                    fp += 1.
                elif score != max([doc.cats['EN'],doc.cats['IS'],doc.cats['ES'],doc.cats['IN']]) and gold[label] < 0.5:
                    tn += 1.
                elif score != max([doc.cats['EN'],doc.cats['IS'],doc.cats['ES'],doc.cats['IN']]) and gold[label] >= 0.5:
                    fn += 1.
            elif label in ['EF','IT','ET','IF']:
                if score == max([doc.cats['EF'],doc.cats['IT'],doc.cats['ET'],doc.cats['IF']]) and gold[label] >= 0.5:
                    tp += 1.
                elif score == max([doc.cats['EF'],doc.cats['IT'],doc.cats['ET'],doc.cats['IF']]) and gold[label] < 0.5:
                    fp += 1.
                elif score != max([doc.cats['EF'],doc.cats['IT'],doc.cats['ET'],doc.cats['IF']]) and gold[label] < 0.5:
                    tn += 1.
                elif score != max([doc.cats['EF'],doc.cats['IT'],doc.cats['ET'],doc.cats['IF']]) and gold[label] >= 0.5:
                    fn += 1.
            elif label in ['EP','IJ','EJ','IP']:
                if score == max([doc.cats['EP'],doc.cats['IJ'],doc.cats['EJ'],doc.cats['IP']]) and gold[label] >= 0.5:
                    tp += 1.
                elif score == max([doc.cats['EP'],doc.cats['IJ'],doc.cats['EJ'],doc.cats['IP']]) and gold[label] < 0.5:
                    fp += 1.
                elif score != max([doc.cats['EP'],doc.cats['IJ'],doc.cats['EJ'],doc.cats['IP']]) and gold[label] < 0.5:
                    tn += 1.
                elif score != max([doc.cats['EP'],doc.cats['IJ'],doc.cats['EJ'],doc.cats['IP']]) and gold[label] >= 0.5:
                    fn += 1.
            elif label in ['NT','SF','ST','NF']:
                if score == max([doc.cats['NT'],doc.cats['SF'],doc.cats['ST'],doc.cats['NF']]) and gold[label] >= 0.5:
                    tp += 1.
                elif score == max([doc.cats['NT'],doc.cats['SF'],doc.cats['ST'],doc.cats['NF']]) and gold[label] < 0.5:
                    fp += 1.
                elif score != max([doc.cats['NT'],doc.cats['SF'],doc.cats['ST'],doc.cats['NF']]) and gold[label] < 0.5:
                    tn += 1.
                elif score != max([doc.cats['NT'],doc.cats['SF'],doc.cats['ST'],doc.cats['NF']]) and gold[label] >= 0.5:
                    fn += 1.
            elif label in ['SP','SJ','NP','NJ']:
                if score == max([doc.cats['SP'],doc.cats['SJ'],doc.cats['NP'],doc.cats['NJ']]) and gold[label] >= 0.5:
                    tp += 1.
                elif score == max([doc.cats['SP'],doc.cats['SJ'],doc.cats['NP'],doc.cats['NJ']]) and gold[label] < 0.5:
                    fp += 1.
                elif score != max([doc.cats['SP'],doc.cats['SJ'],doc.cats['NP'],doc.cats['NJ']]) and gold[label] < 0.5:
                    tn += 1.
                elif score != max([doc.cats['SP'],doc.cats['SJ'],doc.cats['NP'],doc.cats['NJ']]) and gold[label] >= 0.5:
                    fn += 1.
            elif label in ['TJ','TP','FJ','FP']:
                if score == max([doc.cats['TJ'],doc.cats['TP'],doc.cats['FJ'],doc.cats['FP']]) and gold[label] >= 0.5:
                    tp += 1.
                elif score == max([doc.cats['TJ'],doc.cats['TP'],doc.cats['FJ'],doc.cats['FP']]) and gold[label] < 0.5:
                    fp += 1.
                elif score != max([doc.cats['TJ'],doc.cats['TP'],doc.cats['FJ'],doc.cats['FP']]) and gold[label] < 0.5:
                    tn += 1.
                elif score != max([doc.cats['TJ'],doc.cats['TP'],doc.cats['FJ'],doc.cats['FP']]) and gold[label] >= 0.5:
                    fn += 1.
            elif label not in gold:
                continue
                
                
                
    precision = tp / (tp + fp)
    recall = tp / (tp + fn)
    f_score = 2 * (precision * recall) / (precision + recall)
    return {'textcat_p': precision, 'textcat_r': recall, 'textcat_f': f_score}

main(output_dir="./mbti profile model")
```
